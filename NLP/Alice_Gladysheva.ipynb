```python
text=text.replace("\n",".")
text=text.replace('“',".")
text=text.replace('”',".")
text=text.replace(',',".")
text_sentenses=text.split(".",)
sentenses_with_Alice=[]
for i in text_sentenses:
    if("Alice" in i):
        sentenses_with_Alice.append(i)
sentenses_with_Alice=sentenses_with_Alice[4:]        
```

```python
def find_verbs(words):
    verbs=[]
    words=words[0]
    for word in words:
        if("VB" in word[1]):
            verbs.append(word[0])
    return verbs
```

```python
def find_all_verbs(sentence):
    ss=nt.sent_tokenize(sentenses_with_Alice[sentence])
    tokenized_sent=[nt.word_tokenize(sent) for sent in ss]
    pos_sentences=[nltk.pos_tag(sent) for sent in tokenized_sent]
    return find_verbs(pos_sentences)
    
```

```python
all_verbs=[]
for i in range(len(sentenses_with_Alice)):
    all_verbs.append(find_all_verbs(i))
```

```python
verbs=[]
for i in all_verbs:
    if i!=[]:
        for j in i:
            verbs.append(j)
```

```python
verbs=[porter.stem(word) for word in verbs]
```

```python
lst = verbs
dct = {}
def analysis():
    for i in lst:
        if i in dct:
            dct[i] += 1
        else:
            dct[i] = 1  
analysis()
```
